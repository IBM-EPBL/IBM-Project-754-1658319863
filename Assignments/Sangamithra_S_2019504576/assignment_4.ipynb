```python
data=[]
for i in range(0,5572):
    review=dataset["v2"][i]
    review=review.lower()
    review=re.sub('[^a-z]'," ",review)
    review=review.split()
    review=[ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review=' '.join(review)
    data.append(review)
```

```python
data
```

```python
dataset['v1'].value_counts()
```

```python
corpus=" ".join(dataset["v2"])
```

```python
corpus.split()
```

```python
word_count={}

for word in corpus.split():
    if word in word_count:
        word_count[word]= word_count[word]+1
    else:
         word_count[word]=1
        
```

```python
top_15_word=pd.Series(corpus.split()).value_counts().head(15)
```

```python
top_15_word.plot(kind="barh")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
a=cv.fit_transform(data).toarray()
```

```python
a
```

```python
a.shape
```

```python
y=dataset.iloc[:,5:6].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
a_train,a_test,y_train,y_test=train_test_split(a,y,test_size=0.2,random_state=0)
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(units=5572,activation="relu",kernel_initializer="random_uniform"))
```

```python
model.add(Dense(units=3000,activation="relu",kernel_initializer="random_uniform"))
```

```python
model.add(Dense(units=3000,activation="relu",kernel_initializer="random_uniform"))
```

```python
model.add(Dense(units=3000,activation="relu",kernel_initializer="random_uniform"))
```

```python
model.add(Dense(units=1,activation="sigmoid",kernel_initializer="random_uniform"))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
tr=model.fit(a_train,y_train,epochs=20,batch_size=32)
```

```python
loss=tr.history["loss"]
train_accuracy=tr.history["accuracy"]
```

```python
epochs=list(range(1,21))
```

```python
plt.plot(epochs,train_accuracy)
plt.xlabel("epochs")
plt.ylabel("accuracy")
```
